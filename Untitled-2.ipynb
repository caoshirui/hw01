4.A Markov Chain is a probabilistic model where the system transitions from one state to another based on a given set of probabilities. In text generation, the "state" is typically a word or a series of words, and the transition probabilities determine what word comes next.

For a Markovian ChatBot, the model uses the previous word (or series of words) to predict the next word, and the bot generates sentences that follow patterns observed in its training data.

he Markovian ChatBot uses a probabilistic model (Markov Chain) to generate sentences based on training data. The bot breaks the training text into words, builds a chain by mapping each word to the possible next words, and then randomly generates responses by picking words from the chain. The chatbot simulates a conversation with a user by responding with randomly generated text using the Markov chain, but the coherence of the conversation depends heavily on the training data.

```python
import random

# Training data: a string of text to base the Markov Chain on
training_data = "Hello! How are you? I hope you are doing well. It's a nice day, isn't it?"

# Split the training data into words
words = training_data.split()

# Build the Markov Chain
markov_chain = {}

# Loop through the words to create the Markov chain dictionary
for i in range(len(words) - 1):
    current_word = words[i]
    next_word = words[i + 1]
    
    # Add the next word to the list of possible following words for the current word
    if current_word in markov_chain:
        markov_chain[current_word].append(next_word)
    else:
        markov_chain[current_word] = [next_word]

# Function to generate a sentence using the Markov Chain
def generate_sentence(chain, count=10):
    word = random.choice(list(chain.keys()))  # Start with a random word from the keys
    sentence = [word]
    
    # Generate the rest of the sentence
    for i in range(count - 1):
        if word in chain:
            word = random.choice(chain[word])
            sentence.append(word)
        else:
            break  # If the word doesn't exist in the chain, stop
    
    return ' '.join(sentence)

# Chat with the user
while True:
    user_input = input("You: ")
    bot_response = generate_sentence(markov_chain)
    print("Bot:", bot_response)
```

5.1
Character-Specific Markov Chains: Instead of using words as the basis for transitions, we can create Markov chains using individual characters. This allows the chatbot to generate text letter-by-letter, producing more fine-grained and potentially unique outputs. This is useful for generating things like creative names, random text, or simulating typos.
Bigrams (Word-Pair Dependency): A bigram model improves on the basic single-word Markov chain by considering not just the previous word but the previous two words (a pair of words). This adds more context to the predictions, making the generated sentences more coherent and natural compared to the single-word model.

```python
import random

# Training data: some example text
training_data = "Hello! How are you? I hope you are doing well. It's a nice day, isn't it?"

# Build the Markov Chain based on characters
markov_chain = {}

# Loop through the characters to create the Markov chain
for i in range(len(training_data) - 1):
    current_char = training_data[i]
    next_char = training_data[i + 1]
    
    # Add the next character to the list of possible following characters for the current character
    if current_char in markov_chain:
        markov_chain[current_char].append(next_char)
    else:
        markov_chain[current_char] = [next_char]

# Function to generate text using the character Markov Chain
def generate_text(chain, count=100):
    char = random.choice(list(chain.keys()))  # Start with a random character from the keys
    sentence = [char]
    
    # Generate the rest of the sentence
    for i in range(count - 1):
        if char in chain:
            char = random.choice(chain[char])
            sentence.append(char)
        else:
            break  # If the character doesn't exist in the chain, stop
    
    return ''.join(sentence)

# Example usage
print("Generated Text:", generate_text(markov_chain, 100))
```

In the bigram version, the model looks at two words together to predict the next word. This increases the coherence of the responses because it takes the previous two-word context into account rather than just a single word.



```python
import random

# Training data: some example text
training_data = "Hello! How are you? I hope you are doing well. It's a nice day, isn't it?"

# Split the training data into words
words = training_data.split()

# Build the Markov Chain based on bigrams
markov_chain = {}

# Loop through the words to create the bigram Markov chain
for i in range(len(words) - 2):
    current_bigram = (words[i], words[i + 1])
    next_word = words[i + 2]
    
    # Add the next word to the list of possible following words for the current bigram
    if current_bigram in markov_chain:
        markov_chain[current_bigram].append(next_word)
    else:
        markov_chain[current_bigram] = [next_word]

# Function to generate a sentence using the bigram Markov Chain
def generate_sentence(chain, count=10):
    bigram = random.choice(list(chain.keys()))  # Start with a random bigram from the keys
    sentence = list(bigram)  # Convert bigram to a list to start the sentence
    
    # Generate the rest of the sentence
    for i in range(count - 2):
        if bigram in chain:
            next_word = random.choice(chain[bigram])
            sentence.append(next_word)
            bigram = (bigram[1], next_word)  # Update the bigram with the latest word
        else:
            break  # If the bigram doesn't exist in the chain, stop
    
    return ' '.join(sentence)

# Example usage
print("Generated Sentence:", generate_sentence(markov_chain, 10))
```

1.Character-Specific Markov Chain:
This version generates text character by character.
The model builds a Markov chain where each character is linked to the characters that can follow it.
The chatbot generates text by selecting a random starting character and then choosing subsequent characters based on the chain's probabilities.
2.Bigram Model (Word-Pair Dependency):
This version generates text using word pairs (bigrams), which adds more context and coherence to the sentences.
The Markov chain stores word pairs as keys and maps them to the possible words that follow.
The chatbot starts with a random word pair (bigram) and generates the rest of the sentence by predicting the next word based on the bigram.
For both models, the code splits training data into individual characters (for the character-specific chain) or words (for the bigram model), builds a chain by mapping each element to possible next elements, and uses these chains to generate text.

The key differences between the two approaches are the granularity of the transitions (character-level vs. word-pair level) and the level of coherence in the generated sentences.








```python
5.3import random
import re
from collections import defaultdict

class MarkovChainChatBot:
    def __init__(self):
        self.chains = defaultdict(lambda: defaultdict(list))

    def add_corpus(self, character, text):
        # Normalize and tokenize text into bigrams
        words = re.findall(r'\b\w+\b', text.lower())
        bigrams = [(words[i], words[i + 1]) for i in range(len(words) - 1)]
        
        for w1, w2 in bigrams:
            self.chains[character][w1].append(w2)

    def generate_response(self, character, max_length=50):
        if character not in self.chains:
            return "No data for this character."
        
        start_word = random.choice(list(self.chains[character].keys()))
        current_word = start_word
        response = [current_word]

        for _ in range(max_length - 1):
            next_words = self.chains[character].get(current_word, None)
            if not next_words:
                break
            current_word = random.choice(next_words)
            response.append(current_word)

        return ' '.join(response)

# Example usage
if __name__ == "__main__":
    bot = MarkovChainChatBot()

    # Adding character-specific dialogues
    bot.add_corpus("character1", "Hello, how are you doing today? I hope you are well.")
    bot.add_corpus("character2", "What brings you here? It's a lovely day outside.")

    # Generating responses
    print(bot.generate_response("character1"))
    print(bot.generate_response("character2"))
```

Data Structure: A nested dictionary structure (defaultdict) is used to hold bigrams for each character.

Adding Corpus: The add_corpus method normalizes the text and creates bigrams. It then stores the second word of each bigram as a possible continuation from the first word.

Generating Responses: The generate_response method starts with a random word from the chosen character’s chain and constructs a response based on the stored bigrams, up to a specified length.
