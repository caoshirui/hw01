1.To describe the Simple Linear Regression (SLR) model in a straightforward way:

SLR tries to understand the relationship between two variables: the predictor variable (independent variable) and the outcome variable (dependent variable). We model this relationship using a straight line with the following equation:
Y=β0+β1X+ϵ
Here's what each part means:
Y: This is the outcome or dependent variable we are trying to predict.
X: This is the predictor or independent variable that we believe has an impact on 
β0: This is the intercept. It represents the value of 
Y when X is zero.
β1: This is the slope. It tells us how much Y will change for a one-unit change in X.
ϵ: This is the error term. It captures any other factors affecting Y that aren’t explained by X.
In an SLR model, we assume that the errors (ϵ) follow a normal distribution with a mean of 0 and constant variance (σ2). This assumption implies that for each value of 
X, the values of 
Y form a normal distribution around the line 
Y=β0+β1X


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for our model
beta_0 = 2   # Intercept
beta_1 = 0.5 # Slope
sigma = 1    # Standard deviation of the error term

# Generate predictor variable X as random data
np.random.seed(0)
X = np.random.uniform(0, 10, 100)  # 100 random values for X between 0 and 10

# Generate errors following a normal distribution
errors = np.random.normal(0, sigma, 100)

# Calculate the outcome variable Y based on the regression equation
Y = beta_0 + beta_1 * X + errors

# Plotting the results
plt.scatter(X, Y, label="Generated data points", color="blue")
plt.plot(X, beta_0 + beta_1 * X, color="red", label="True regression line")
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Simple Linear Regression Model Simulation")
plt.legend()
plt.show()
```

2.To fit a Simple Linear Regression model on a simulated dataset and visualize the results, we'll go through these steps:
Generate Data: We'll use the same equation 
Y=β0+β1X+ϵ
Create a DataFrame: Use pandas to organize this data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Step 1: Generate data following the SLR model
np.random.seed(0)
beta_0 = 2
beta_1 = 0.5
sigma = 1
X = np.random.uniform(0, 10, 100)
errors = np.random.normal(0, sigma, 100)
Y = beta_0 + beta_1 * X + errors

# Step 2: Create a DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Step 3: Fit a Simple Linear Regression model
model = smf.ols(formula='Y ~ X', data=data)
results = model.fit()

# Print the summary of the regression model
print(results.summary())

# Step 4: Visualize the data and the fitted regression line
plt.scatter(data['X'], data['Y'], label="Generated data points", color="blue")
plt.plot(data['X'], results.predict(data['X']), color="red", label="Fitted regression line")
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Fitted Simple Linear Regression Model")
plt.legend()
plt.show()
```

5.The "ordinary least squares" (OLS) method chooses a line that minimizes the sum of the squared residuals (differences between observed and predicted values) based on observed data.

Residuals: In the fitted model, each observed Y value has a predicted value, Yi^.The difference between the observed Yi^ and (predicted Y) is the residual, ei=Yi− Yi^.
Minimizing Squared Residuals: OLS finds the line that minimizes the sum of ei^2(the squared residuals). Squaring is necessary because residuals can be positive or negative, and squaring ensures they contribute equally to the total "error" regardless of sign. Squaring also penalizes larger residuals more heavily, prioritizing a line that closely fits all data points.

Purpose of the Fit: By minimizing the sum of squared residuals, OLS produces the line that best represents the trend in the observed data, balancing out the variability in a way that minimizes the overall prediction error.

In the visualization, the red dashed lines show these residuals, demonstrating how the fitted line (blue) is chosen to make these residuals as small as possible in the squared sense.

7.Linearity: The relationship between the predictor x and outcome Y should be linear. If the data shows a non-linear pattern (e.g., curving or cyclical trends), it violates this assumption, indicating that the SLR model may not fit well.

Homoscedasticity (Constant Variance of Errors): The variance of the residuals (errors) should be constant across all values of the predictor x. If the residuals fan out (increasing variance) or funnel in (decreasing variance) as x changes, this suggests heteroscedasticity, violating the SLR assumption.

Normality of Errors: The residuals should follow a normal distribution. If the residuals have a skewed distribution or show significant outliers, this assumption is violated, which affects hypothesis tests and confidence intervals.

Independence of Errors: The errors should be independent of each other. If there’s evidence of autocorrelation (especially in time-series data), this assumption doesn’t hold.

13.(A) Permutation Test
The permutation test involves "shuffling" the labels between the short and long groups to simulate the null hypothesis of no difference in means. If the null hypothesis is true, the observed difference in means should be similar to the distribution of mean differences from shuffled labels.

Null Hypothesis (H0): μ short =μ long , meaning there is no difference in duration between short and long wait groups.
Procedure:
Shuffle the labels of the "kind" column in the dataset so that each duration is randomly assigned to either "short" or "long," breaking any association between duration and wait type.
Calculate the mean difference for each shuffle.
Repeat this shuffling many times (e.g., 1,000 permutations) to create a distribution of mean differences under the null hypothesis.
Compare the observed mean difference (calculated from the original data) to this distribution to get a p-value, which represents how likely it is to observe a difference as extreme as the observed one under the null hypothesis.

```python
import numpy as np
import pandas as pd

# Observed mean difference
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Permutation test
n_permutations = 1000
permuted_diffs = []

for _ in range(n_permutations):
    # Shuffle the labels
    shuffled = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(frac=1).values)
    # Calculate mean difference
    perm_diff = shuffled.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permuted_diffs.append(perm_diff)

# Calculate p-value
p_value = np.mean(np.abs(permuted_diffs) >= np.abs(observed_diff))
print(f"Permutation Test p-value: {p_value}")
```

13.(B) Bootstrap Confidence Interval
The bootstrap confidence interval method resamples each group separately with replacement to estimate the difference in means between the "short" and "long" groups.

Procedure:
Draw a bootstrap sample for each group (short and long) with replacement.
Calculate the mean for each group and then compute the difference between the two means.
Repeat this process (e.g., 1,000 bootstrap samples) to get a distribution of mean differences.
Compute the 2.5th and 97.5th percentiles of this distribution to form a 95% confidence interval.

```python
# Bootstrap Confidence Interval
n_bootstrap = 1000
bootstrapped_diffs = []

for _ in range(n_bootstrap):
    # Resample within each group
    boot_sample = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    # Calculate mean difference
    boot_diff = boot_sample.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrapped_diffs.append(boot_diff)

# Calculate the 95% confidence interval
ci_lower, ci_upper = np.quantile(bootstrapped_diffs, [0.025, 0.975])
print(f"95% Bootstrap Confidence Interval for Mean Difference: [{ci_lower}, {ci_upper}]")
```

Bootstrap Confidence Interval: By resampling each group independently, we generate an empirical distribution of mean differences. This approach does not assume the null hypothesis and instead estimates the variability in the difference directly, allowing us to build a confidence interval.
